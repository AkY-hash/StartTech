### Converting Categorial Variables to Numerical Variables

```python
#Converting by using get_dummies

train_data =pd.get_dummies(train_data, drop_first=True)
train_data.head()
```

```python
test_data = pd.get_dummies(test_data, drop_first=True)
test_data.head()
```

### Training Linear Regression Model

```python
#importing Linear Regression from the sklearn library

from sklearn.linear_model import LinearRegression 
reg = LinearRegression()
```

```python
x = train_data.drop("CTC", axis = 1)
y = train_data["CTC"]
```

```python
#fitting the data

reg.fit(x,y)
```

```python
#predicting data

y_pred = reg.predict(x)
print("Predictions:", y_pred)
```

### Evaluating Results

```python
#importing Mean Squared Error Metric from sklearn library

from sklearn.metrics import mean_squared_error
```

```python
#calculating and printing Mean Squared Error

print('Mean Squared Error:', mean_squared_error(y, y_pred))
```

### Predicting

```python
#predicting the testing model

predicted = pd.concat([test_data, pd.DataFrame(y_pred, columns=["Predicted CTC"])], axis=1)
```

```python
predicted = predicted[["College_Tier 2", "College_Tier 3", "City type_Non-Metro", "Graduation marks", "Role_Manager", "Actual CTC", "Predicted CTC"]]
```
